# StackAPI

#### Import the necessary libraries here:

```python
# your code here
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
```

#### Question 1: Find the questions and answers of last month.

```python
# your code here
questions = SITE.fetch('questions', fromdate=1580515200, todate=1583020799)
answers = SITE.fetch('answers', fromdate=1580515200, todate=1583020799)
```

```python
print(questions)
print(answers)
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
high_questions = SITE.fetch('questions', fromdate=1583625600, todate=1583688015,
    min=5, tagged='python', sort='votes')
high_questions
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
# your code here
answers_specific = SITE.fetch('answers/{ids}', ids=[6784, 6473])
answers_specific
```
