#### 1. Obtain the full list of forks created from the main lab repo via Github API.

To list forks, we can use the GET method. As explained in the GitHub API documentation, we need to make the request to: GET /repos/:owner/:repo/forks.

`````python
# your code here
username = 'blabla'
token = 'blablabla'

response = requests.get('https://api.github.com/repos/ironhack-datalabs/dataV2-labs/forks', auth=(username,token))
data = response.json()
`````

#### 2. Loop the JSON response to find out the language attribute of each fork. Use an array to store the language attributes of each fork.
Hint: Each language should appear only once in your array.

`````python
flattened_data.columns
`````

`````python
language_url = list(flattened_data["languages_url"])
language_url
`````

`````python
# your code here

language_response = []

for url in language_url:
    x = requests.get(url, auth=(username,token)).json()
    language_response.append(x)   
`````

#### 3. Read the content of each secret files into an array of strings.
Since the response is encoded, you will need to send the following information in the header of your request:
````python
headers = {'Accept': 'application/vnd.github.v3.raw'}
````

`````python
#I now know the filenames, need to find corresponding paths.
scavenger_url = [val['url'] for sublist in files for val in sublist]
scavenger_url

scavenger_url = [file for file in scavenger_url if 'scavengerhunt' in file]
scavenger_url.sort()
scavenger_url

sorted_url = sorted(scavenger_url, key = lambda x: x.split(".")[3])
sorted_url
`````

`````python
# your code here

headers = {'Accept': 'application/vnd.github.v3.raw'}

joke_response = []

for url in sorted_url:
    x = requests.get(url, headers=headers, auth=(username,token)).text
    joke_response.append(x) 

joke_response
`````

#### 4. Concatenate the strings in the array separating each two with a whitespace.

`````python
# your code here

joke = [i.strip() for i in joke_response]
joke = ' '.join(joke)
`````
