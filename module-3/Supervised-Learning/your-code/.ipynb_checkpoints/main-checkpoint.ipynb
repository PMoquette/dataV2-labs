Too many unique values in a column is not desirable in machine learning because it makes prediction inaccurate. But there are workarounds under certain conditions. One of the fixable conditions is:

The `WHOIS_COUNTRY` column happens to be this case. You can verify it by print a bar chart of the `value_counts` in the next cell to verify:

```python
# Your code here
fig, ax = plt.subplots(figsize=(15,10))
plt.xticks(rotation=90)
bar_plot = sns.barplot(x=clean_websites.WHOIS_COUNTRY.value_counts().index, y=clean_websites.WHOIS_COUNTRY.value_counts(), data=clean_websites)
```

```python
clean_websites['WHOIS_COUNTRY'].value_counts()
```

#### After verifying, now let's keep the top 10 values of the column and re-label other columns with `OTHER`.

```python
# Your code here
# Top 10 are US, None, CA, ES, AU, PA, GB, JP, UK, CN 
# First thing's first: there are three different values that all refer to the UK. These need to be changed first.
# Then I can assign all values that are not in the list above to 'Other'
clean_websites['WHOIS_COUNTRY'] = np.where((clean_websites['WHOIS_COUNTRY'] == "[u'GB'; u'UK']"),'UK',clean_websites['WHOIS_COUNTRY'])
clean_websites['WHOIS_COUNTRY'] = np.where((clean_websites['WHOIS_COUNTRY'] == "United Kingdom"),'UK',clean_websites['WHOIS_COUNTRY'])
clean_websites['WHOIS_COUNTRY'].value_counts()
```

```python
countries = ['US', 'None', 'CA', 'ES', 'AU', 'PA', 'GB', 'JP', 'UK', 'CN']

clean_websites.loc[~clean_websites['WHOIS_COUNTRY'].isin(countries), 'WHOIS_COUNTRY']= 'Other'

#df.loc[~df["S"].isin(allowed_vals), "S"] = "None"
```

Although there are so many unique values in the `SERVER` column, there are actually only 3 main server types: `Microsoft`, `Apache`, and `nginx`. Just check if each `SERVER` value contains any of those server types and re-label them. For `SERVER` values that don't contain any of those substrings, label with `Other`.

At the end, your `SERVER` column should only contain 4 unique values: `Microsoft`, `Apache`, `nginx`, and `Other`.

```python
# Your code here
# https://stackoverflow.com/questions/26577516/how-to-test-if-a-string-contains-one-of-the-substrings-in-a-list-in-pandas
# clean_websites['SERVER'][clean_websites['SERVER'].str.contains('|'.join(servers))]
import re
servers = ['microsoft', 'apache', 'nginx']
pattern = re.compile(rf"({'|'.join(servers)})")

def try_extract(s):
    try:
        return pattern.search(s.lower()).group(1)
    except (TypeError, AttributeError):
        return 'Other'

clean_websites['SERVER'] = [try_extract(s) for s in clean_websites['SERVER']]
```

```python
# Count `SERVER` value counts here
clean_websites['SERVER'].unique()
clean_websites['SERVER'].value_counts()
```

OK, all our categorical data are fixed now. **Let's convert them to ordinal data using Pandas' `get_dummies` function ([documentation](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.get_dummies.html)).** Make sure you drop the categorical columns by passing `drop_first=True` to `get_dummies` as we don't need them any more. **Also, assign the data with dummy values to a new variable `website_dummy`.**
