# Inbalanced Classes
## In this lab, we are going to explore a case of imbalanced classes. 


Like we disussed in class, when we have noisy data, if we are not careful, we can end up fitting our model to the noise in the data and not the 'signal'-- the factors that actually determine the outcome. This is called overfitting, and results in good results in training, and in bad results when the model is applied to real data. Similarly, we could have a model that is too simplistic to accurately model the signal. This produces a model that doesnt work well (ever).

```python
paysim.drop(['oldbalanceDest','oldbalanceOrg'], axis=1, inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(15,10)) 

sns.heatmap(paysim.corr(), cmap='coolwarm', annot=True, vmax=1, vmin=-1, center=0,
            linewidths=.5)

bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
#nameDest and nameOrig is customer or recipient ID for transaction. C is customer, M is merchant. In the latter case, no information is given. 
#this makes 0 values for new/old values where the customer is M not necessarily meaningful (it isn't really 0, the information is just missing)
# nameOrig/nameDest
paysim["classOrig"]= paysim["nameOrig"].str.get(0)
paysim["classDest"]= paysim["nameDest"].str.get(0)
```

```python
paysim.head()
```

```python
#the name ID's themselves are too many variables to take into account, so drop those.
paysim.drop(['nameOrig','nameDest'], axis=1, inplace=True)
```

```python
# New codebook: C=1, M=0
paysim['classOrig'] = np.where((paysim.classOrig == 'C'),1,0)
paysim['classDest'] = np.where((paysim.classDest == 'C'),1,0)
```

```python
paysim.head()
```

```python
#Ok so the time-thing. From the data description on Kaggle:
# Step maps a unit of time in the real world. In this case 1 step is 1 hour of time. Total steps 744 (30 days simulation).
# We need time for resampling:
import datetime 
start_datetime =  datetime.datetime(2020, 4, 1, 0,0) 
paysim['datetime'] = paysim.apply(lambda x: start_datetime + datetime.timedelta(hours=x['step']), axis=1)
paysim.drop(['step'], axis=1, inplace=True)
```

```python
paysim.set_index('datetime',inplace=True)
paysim.reset_index(inplace=True)
```

```python
paysim.head()
```

```python
dummies = pd.get_dummies(paysim['type'])
paysim_dummies = pd.concat([paysim, dummies], axis=1)
```

```python
paysim_dummies.drop(['type','isFlaggedFraud'], axis=1, inplace=True)
```

```python
paysim_dummies.shape
```

```python
fig, ax = plt.subplots(figsize=(15,10)) 

sns.heatmap(paysim_dummies.corr(), annot=True, cmap='coolwarm', linewidth=0.5)

bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
paysim_dummies['classOrig'].value_counts() # They're all C..and classDest is very highly correlated with Payment, so drop both after all
paysim_dummies.drop(['classOrig','classDest'], axis=1, inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(15,10)) 

sns.heatmap(paysim_dummies.corr(), annot=True, cmap='coolwarm', linewidth=0.5)

bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

### Run a logisitc regression classifier and evaluate its accuracy.

```python
# Your code here
# 0    6354407
# 1       8213
from sklearn.utils import resample
majority = paysim_dummies[paysim_dummies['isFraud'] == 0]
minority = paysim_dummies[paysim_dummies['isFraud'] == 1]
 
# Downsample majority class
majority_downsampled = resample(majority, replace=False, n_samples=8213, random_state=123)

# Combine minority class with downsampled majority class
paysim_downsampled = pd.concat([majority_downsampled, minority])

paysim_downsampled['isFraud'].value_counts()
```

```python
paysim_downsampled.head()
```

### Now pick a model of your choice and evaluate its accuracy.

```python
# Your code here
from sklearn.model_selection import train_test_split

y = paysim_downsampled['isFraud']
x = paysim_downsampled.drop(['isFraud','datetime'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.fit_transform(X_test)

```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver='liblinear', random_state=0)
logreg.fit(scaled_X_train, y_train)

y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

print(confusion_matrix(y_test, y_pred))
print(precision_score(y_test, y_pred))
print(recall_score(y_test, y_pred))
print(f1_score(y_test, y_pred))
```

### Which model worked better and how do you know?

```python
# Your response here
from sklearn.metrics import roc_curve

precision, recall,_ = roc_curve(y_test, y_pred)

plt.plot(precision, recall, marker='o', label='Logistic_Regression')

plt.legend()

plt.show()
```
