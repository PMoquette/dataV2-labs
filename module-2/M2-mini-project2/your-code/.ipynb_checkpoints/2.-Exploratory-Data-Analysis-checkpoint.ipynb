### Looking at the distributions of the data
Let's take a look at the distribution of the 4 stats that describe the physical characteristics of the players.

**Plot the four distributions about `height`, `weight`, `age` and `BMI`.**

```python
#your code here
sns.set(style="whitegrid", palette="muted", font_scale=2, color_codes=True)
f, axes = plt.subplots(2, 2, figsize=(25, 25), sharex=False)

sns.distplot(wnba['Height'], color="b", ax=axes[0, 0])

# Plot a kernel density estimate and rug plot
sns.distplot(wnba['Weight'], color="r", ax=axes[0, 1])

# Plot a filled kernel density estimate
sns.distplot(wnba['Age'], color="g", kde_kws={"shade": True}, ax=axes[1, 0])

# Plot a histogram and kernel density estimate

sns.distplot(wnba['BMI'], color="y", ax=axes[1, 1])

plt.setp(axes, yticks=[])
plt.tight_layout()

'''
sns.distplot(wnba['Height'])
sns.distplot(wnba['Weight'])
sns.distplot(wnba['Age'])
sns.distplot(wnba['BMI'])
'''
```

**What conclusions do you think we can take from this plots?**

As the three fields (Height, Weight, BMI) are all dependent on each other, it makes sense that these are shaped more or less the same, with the BMI a bit more towards a normal distribution (smoother). Age is left skewed, with a higher amount of younger (below 30 y/o) players, which makes sense in professional sports. 

Height is bimodal, two groups/preferences for different positions/roles?

In addition to what the describe() method already told us we can see that the physical characteristics of the players (apart from the age) more or less follow a standard distribution, which is expected when looking at the distribution of values that arise from a lot of factors that are independent from each other as is the case for many physical quantities. 

The height distribution looks like a bimodal. This may be due to the fact that players in basketball fall under two main categories (please note that this is a very gross generalization): shorter and more agile, and taller and less agile. Therefore there are less "average" height players since they will neither be as agile as the shorter players nor have the same impact in the paint (that is, under the basket) as a taller player.

The age distribution is a bit skewed to the right which is expected since most professional players stop playing after their prime physical years come to an end.

We could do the same for the main game stats. They are points, assists, blocks, rebounds and steals. 

**Now plot the distribution of the columns `REB`, `AST`, `STL`, `PTS` and `BLK` the same way you did in the last cell.**

```python
#your code here
sns.set(style="whitegrid", palette="deep", font_scale=2, color_codes=True)
f, axes = plt.subplots(3, 2, figsize=(25, 25), sharex=False)

sns.distplot(wnba['REB'], color="b", ax=axes[0, 0])

# Plot a kernel density estimate and rug plot
sns.distplot(wnba['AST'], color="r", ax=axes[0, 1])

# Plot a filled kernel density estimate
sns.distplot(wnba['STL'], color="g", kde_kws={"shade": True}, ax=axes[1, 0])

# Plot a histogram and kernel density estimate

sns.distplot(wnba['PTS'], color="y", ax=axes[1, 1])

sns.distplot(wnba['BLK'], color="purple", ax=axes[2, 0])

plt.setp(axes, yticks=[])
plt.tight_layout()
```

**What conclusions do you think we can take from this plots?**

|REB | Total Rebounds<br>
| AST | Assists<br>
| STL | Steals<br>
| BLK | Blocks<br>
| PTS | Total points<br> 

These are all left skewed (need to fix the x axis, but not now. Can't have negative values for these scores). It makes sense that there are so many lower values; not every player plays as much, not every team member is in a position where they can make these moves. Again, have no clue about basketball, but I imagine it's like football; not everyone makes it to the front of the field. Some are more support, others defense, others attack. 

As expected all of the above distribution are heavily skewed to the right, since most players will have very low to average stats while there will be a handful of star players whose stats peak above everyone else. It is also important to think about the fact that we are simply taking the stats as they are without considering the minutes played by each player. Even though skill plays a very important factor in determining this kind of stats we also have to consider that players that play more minutes will, on average, score more points (or blocks, assists, etc.).

**For the sake of it let's look at the same distributions by dividing those stats by the minutes played for each player in the dataset.** 

```python
#your code here
new_df = wnba[['Height', 'Weight', 'Age', 'BMI', 'REB', 'AST', 'STL', 'PTS', 'BLK', 'MIN']]
new_df

new_df1 = new_df[['Height', 'Weight', 'Age', 'BMI', 'REB', 'AST', 'STL', 'PTS', 'BLK']].div(new_df['MIN'],axis=0)
new_df1
```

```python
df_players = new_df1.filter(['Height', 'Weight', 'Age', 'BMI'])

df_points = new_df1.filter(['REB', 'AST', 'STL', 'PTS', 'BLK'])
```

```python
'''sns.set(style="whitegrid", font_scale=1.5)

f, axes = plt.subplots(3, 3, figsize=(20, 20), sharex=False)

ax=axes[0,0]
ax=axes[0,1]
ax=axes[1,0]
ax=axes[1,1]
ax=axes[0,2]
ax=axes[2,0]
ax=axes[2,1]
ax=axes[1,2]
ax=axes[2,2]

for i, column in enumerate(new_df1.columns):
    sns.distplot(new_df1[column],ax=axes[i//3,i%3])

plt.setp(axes, yticks=[])
plt.tight_layout()
'''
sns.set(style="whitegrid", font_scale=1.5)
f, axes = plt.subplots(2, 2, figsize=(20, 20), sharex=False)

ax=axes[0,0]
ax=axes[0,1]
ax=axes[1,0]
ax=axes[1,1]


for i, column in enumerate(df_players.columns):
    sns.distplot(new_df1[column],ax=axes[i//2,i%2])

plt.tight_layout()
```

```python
f, axes = plt.subplots(3, 2, figsize=(20, 20), sharex=False)

ax=axes[0,0]
ax=axes[0,1]
ax=axes[1,0]
ax=axes[1,1]
ax=axes[2,1]


for i, column in enumerate(df_points.columns):
    sns.distplot(new_df1[column],ax=axes[i//2,i%2])

plt.tight_layout()
```

**What conclusions do you think we can take from this plots?**

We can see that there are certain typical ranges for Age, Weight, Height, and BMI (stats my sister would be interested to know), but they are less restrictive than one may think, as there are many different positions and roles within a team.
